<H2> Parametric Curves in 3D </H2>
<p> Let's begin with plotting the line segment joining the points (1,2,2) and (3,6,1). The line segment has a vector parametric represenation r(t)=<1+2t,2+4t,2-t>, with 0 &#8804 t &#8804 1. </p>

```python
ax = plt.figure(figsize=(16,14)).add_subplot(projection='3d')

# Prepare arrays x, y, z
t=np.linspace(0, 1, 10)
x = 1+2*t
y = 2+4*t
z=  2-t

#plot
ax.plot(x, y, z, label='Line segment 1')
ax.legend()

plt.show()
```


<p> We can extend the line segment above extending the parameter's domain to the interval -50 &#8804 t &#8804 50. This would give you the line segment joining the points (-99,-198, 52) and (101, 202, -48)</p>

```python
ax = plt.figure(figsize=(16,14)).add_subplot(projection='3d')


t=np.linspace(-50, 50, 10)
x = 1+2*t
y = 2+4*t
z=  2-t

ax.plot(x, y, z, label='Line segment 2', color='green')
ax.legend()

plt.show()
```

<p> Here we plot the parametric curve &lt cos(t),t,sin(t) &gt, with -50 &#8804 t &#8804 50.</p>

```python
ax = plt.figure(figsize=(16,14)).add_subplot(projection='3d')


t=np.linspace(-50, 50, 1000)
x = np.cos(t)
y = t
z=  np.sin(t)
ax.plot(x, y, z, label='Parametric curve 3', color='darkblue')
ax.legend(loc='upper right', bbox_to_anchor=(1.0, 1.0))

plt.show()
```

<p> Next we project the previous curve onto the three coordinate planes. </p>

```python
ax1 = plt.figure(figsize=(6,10)).add_subplot()


t=np.linspace(-50, 50, 1000)
x = np.cos(t)
y = t
ax1.plot(x, y, label='Projection of the parametric curve 3 onto xy-plane')
ax1.legend(loc='upper right', bbox_to_anchor=(1.0, 1.1))

ax2 = plt.figure(figsize=(8,8)).add_subplot()


t=np.linspace(-50, 50, 1000)
x = np.cos(t)
z=  np.sin(t)
ax2.plot(x,z, label='Projection of the parametric curve 3 onto xz-plane', color='darkgreen')
ax2.legend(loc='upper right', bbox_to_anchor=(1.0, 1.1))

ax3 = plt.figure(figsize=(8,6)).add_subplot()


t=np.linspace(-50, 50, 1000)
y = t
z=  np.sin(t)
ax3.plot(y, z, label='Projection of the parametric curve 3 onto yz-plane', color='purple')
ax3.legend(loc='upper right', bbox_to_anchor=(1.0, 1.1))


plt.show()
```

<p> We plot now the curve with parametric equations &lt cos(t),t, sin(t)/t &gt, with 0 &lt t &#8804 50.</p>

```python
ax = plt.figure(figsize=(16,14)).add_subplot(projection='3d')


t=np.linspace(0.0000001, 50, 1000)
x = np.cos(t)
y = t
z=  (np.sin(t))/t
ax.plot(x, y, z, label='Parametric curve 6', color='darkblue')
ax.legend(loc='upper right', bbox_to_anchor=(1.0, 1.0))

plt.show()
```

<p> As before, we project the previous curve onto the three coordinate planes. </p>

```python
ax1 = plt.figure(figsize=(6,10)).add_subplot()


t=np.linspace(0.000001, 50, 1000)
x = np.cos(t)
y = t
ax1.plot(x, y, label='Projection of the parametric curve 6 onto xy-plane')
ax1.legend(loc='upper right', bbox_to_anchor= (1.0, 1.1))



ax2 = plt.figure(figsize=(8,8)).add_subplot()


t=np.linspace(0.0000001, 50, 1000)
x = np.cos(t)
z=  (np.sin(t))/t
ax2.plot(x,z, label='Projection of the parametric curve 6 onto xz-plane', color='darkgreen')
ax2.legend(loc='upper right', bbox_to_anchor=(1.0, 1.1))

ax3 = plt.figure(figsize=(10,6)).add_subplot()


t=np.linspace(0.0000001, 50, 1000)
y = t
z=  (np.sin(t))/t
ax3.plot(y, z, label='Projection of the parametric curve 6 onto yz-plane', color='purple')
ax3.legend(loc='upper right', bbox_to_anchor=(1.0, 1.1))


plt.show()
```

<p> Just for fun! We can plot serveral curves together.</p>

```python
ax = plt.figure(figsize=(18,16)).add_subplot(projection='3d')

# Prepare arrays x, y, z
t=np.linspace(-125, 125, 1000)
s=np.linspace(-2,2,10)
r=np.linspace(-5,5,1000)
x1 = np.cos(t) 
y1 = np.sin(t)
z1=  t
x2 = np.sin(r) 
y2 = np.cos(r)
z2=  r**3
x3 = 0+s 
y3 = 0+2*s
z3=  0+5*s
ax.plot(x1, y1, z1, label='parametric curves 1')
ax.plot(x2, y2, z2, label='parametric curves 2', color='red')
ax.plot(x3, y3, z3, label='parametric curves 3')
ax.legend()
plt.show()
```

<p> We can now plot some surfaces, starting with the plane x+y+z=0.</p>

```python

ax = plt.figure().add_subplot(projection='3d')

# Prepare arrays x, y, z
X = np.linspace(-10,10,500)
Y = np.linspace(-10,10,500)
X, Y = np.meshgrid(X, Y)
Z=-X-Y
ax.plot_surface(X, Y, Z, label='Plane x+z+y=0')
#ax.legend()

plt.show()
```

```python
ax1 = plt.figure(figsize=(8,6)).add_subplot()


x=np.linspace(-10, 10, 500)
y = -x
ax1.plot(x, y, label='Trace of the plane x+y+z=0 onto xy-plane')
ax1.legend(loc='upper right', bbox_to_anchor=(1.5, 1.1))

ax2 = plt.figure(figsize=(8,6)).add_subplot()


x=np.linspace(-10, 10, 500)
z= -x
ax2.plot(x,z, label='Trace of the plane x+y+z=0 onto xz-plane', color='darkgreen')
ax2.legend(loc='upper right', bbox_to_anchor=(1.5, 1.1))

ax3 = plt.figure(figsize=(8,6)).add_subplot()


y=np.linspace(-10, 10, 500)
z=  -y
ax3.plot(y, z, label='Trace of the plane x+y+z=0 onto yz-plane', color='purple')
ax3.legend(loc='upper right', bbox_to_anchor=(1.5, 1.1))


plt.show()
```

<p> Surface 1 : z<sup>2</sup>+3x<sup>2</sup>+25y<sup>2</sup>=10 </p>

```python
ax = plt.figure(figsize=(10,8)).add_subplot(projection='3d')

# Prepare arrays x, y, z
X = np.linspace(-2,2,3000)
Y = np.linspace(-2,2,3000)
X, Y = np.meshgrid(X, Y)

equation1=-3*X**2-25*Y**2+10
Z=np.sqrt(equation1)
ax.plot_surface(X, Y, Z, color='blue')

Z1=- np.sqrt(equation1)
ax.plot_surface(X, Y, Z1, color='blue')

plt.show()
```

<p> Traces of surface 1</p>

```python
ax1 = plt.figure(figsize=(8,6)).add_subplot()

x=np.linspace(-2, 2, 3000)
equation2=-(3/25)*x**2+(2/5)
y = np.sqrt(equation2)
y1=-np.sqrt(equation2)
ax1.plot(x, y, label='Trace of surface 1 onto xy-plane', color='blue')
ax1.plot(x, y1, color='blue')
ax1.legend(loc='upper right', bbox_to_anchor=(1.5, 1.1))



ax2 = plt.figure(figsize=(6,8)).add_subplot()

x=np.linspace(-2, 2, 3000)
equation2=-3*x**2+10
z = np.sqrt(equation2)
z1=-np.sqrt(equation2)
ax2.plot(x, z, label='Trace of surface 1 onto xz-plane', color='red')
ax2.plot(x, z1, color='red')
ax2.legend(loc='upper right', bbox_to_anchor=(1.5, 1.1))


ax2 = plt.figure(figsize=(6,8)).add_subplot()

y=np.linspace(-2, 2, 3000)
equation2=-25*y**2+10
z = np.sqrt(equation2)
z1=-np.sqrt(equation2)
ax2.plot(y, z, label='Trace of surface 1 onto yz-plane', color='green')
ax2.plot(y, z1, color='green')
ax2.legend(loc='upper right', bbox_to_anchor=(1.5, 1.1))


plt.show()
```
