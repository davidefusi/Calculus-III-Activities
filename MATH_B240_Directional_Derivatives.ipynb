<b> Example </b>

<p> For the function in two variable f(x,y)=x<sup>2</sup>+y<sup>2</sup>, we fix the point (2,0,4) on the graph z=f(x,y). In the graph below, you can see the surface z=x<sup>2</sup>+y<sup>2</sup>, and the curves C1, C2, and C3 determined when moving along the lines: L1=<2+t,0>, L2=<2,t> and L3=<2+(3/5)t,(4/5)t> in the xy-plane.  </p>

```python
ax = plt.figure(figsize=(10,8)).add_subplot(projection='3d')

# Prepare arrays x, y, z
X = np.linspace(-4,4,100)
Y = np.linspace(-4,4,100)
X, Y = np.meshgrid(X, Y)




t=np.linspace(-4, 4, 100)
s=np.linspace(-4,4,100)
r=np.linspace(-4,4,100)
x1 = 2+t 
y1 = 0*t
z1=  (2+t)**2
#x1p = 2+t 
#y1p = 0*t
#z1p =  0*t
x2 = 2+0*s
y2 = s
z2=  4+(s)**2
x3 = 2+(3/5)*r
y3 = (4/5)*r
z3=  (2+(3/5)*r)**2+((4/5)*r)**2
ax.plot(x1, y1, z1, label='Curve C1 along line L1', color='darkblue',zorder=3)
ax.plot(x2, y2, z2, label='Curve C2 along line L2', color='darkgreen',zorder=3)
ax.plot(x3, y3, z3, label='Curve C3 along line L3', color='red',zorder=3)
#ax.plot(x1p, y1p, z1p, label='Curve C3 along line L3', color='red',zorder=3)
ax.legend(loc='upper right', bbox_to_anchor=(1.5, 1.0))
Z=X**2+Y**2
ax.plot_surface(X, Y, Z, color='white')

plt.show()
```

<p> The curves C1, C2, and C3 are plane curves contained in the planes y=0, x=2, and x-(3/4)y=2 respectively. We can look each curve in the vertical plane the curve belongs to. We can also draw the line tangent to the curve at the point (2,0,4).</p>

```python
ax1 = plt.figure(figsize=(6,10)).add_subplot()


t=np.linspace(-4, 4, 100)
x = t+2
y = (2+t)**2
ax1.plot(x, y, label='Curve C1 in the plane y=0', color='darkblue')
s=np.linspace(-4, 4, 100)
w=2+s
v=4+4*s
ax1.plot(w, v, label='Line tangent to C1 at (2,0,4)',color='lightblue')
ax1.legend(loc='upper right', bbox_to_anchor=(1.6, 1.0))

ax2 = plt.figure(figsize=(8,8)).add_subplot()


t=np.linspace(-4, 4, 100)
x = t
z=  4+(t)**2
ax2.plot(x,z, label='Curve C2 in the plane x=2', color='darkgreen')
s=np.linspace(-4, 4, 100)
w=s
v=4+s*0
ax2.plot(w, v, label='Line tangent to C2 at (2,0,4)',color='lightgreen')
ax2.legend(loc='upper right', bbox_to_anchor=(1.6, 1.0))

ax3 = plt.figure(figsize=(8,6)).add_subplot()


t=np.linspace(-4, 4, 100)
x = t+2
z=  (2+(3/5)*t)**2+((4/5)*t)**2
ax3.plot(x, z, label='Curve C3 in the plane x-(3/4)y=2', color='red')
s=np.linspace(-4, 4, 100)
w=2+s
v=4+s*(12/5)
ax3.plot(w, v, label='Line tangent to C3 at (2,0,4)',color='pink')
ax3.legend(loc='upper right', bbox_to_anchor=(1.6, 1.0))


plt.show()
```

<p> Next, we can visualize the curves and the corresponding tangent lines in the same plane. For example, we can rotate the planes x=2 and x-(3/4)y=2 along the vertical line L=<2,0,t> (or x-2=y=0) so to coincide with the plane y=0. </p>

```python
ax1 = plt.figure(figsize=(6,10)).add_subplot()


t=np.linspace(-4, 4, 100)
s=np.linspace(-4, 4, 100)

x1 = t+2
y1 = (2+t)**2
ax1.plot(x1, y1, color='darkblue', label="Curve C1")
w1=2+s
v1=4+4*s
ax1.plot(w1, v1, color='lightblue', label="Line tanget to C1")


x2 = t
z2=  4+(t-2)**2
ax1.plot(x2,z2, color='darkgreen', label="Curve C2")
w2=s
v2=4+s*0
ax1.plot(w2, v2, color='lightgreen', label="Line tanget to C2")


x3 = t+2
z3=  (2+(3/5)*t)**2+((4/5)*t)**2
ax1.plot(x3, z3, color='red',label="Curve C3")
w3=2+s
v3=4+s*(12/5)
ax1.plot(w3, v3,color='pink', label="Line tanget to C1")

ax1.legend(loc='upper right', bbox_to_anchor=(1.6, 1.0))
plt.title('All curves and tangent lines on the same plane')
plt.show()
```

<p> <b> Remembering the concepts seen in class</b>, we can now recognize that: the slope of the line tangent to the curve C1 is the partial derivative of f(x,y) with respect to x at the point (2,0); the slope of the line tangent to the curve C2 is the partial derivative of f(x,y) with respect to y at the point (2,0); and that the slope of the line tangent to the curve C3 is the directional derivative of f(x,y) at the point (2,0) in the direction <3/5, 4/5>. </p>
